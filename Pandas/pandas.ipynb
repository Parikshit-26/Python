## What is pandas :
A powerful library for data manipulation and analysis in python. Built on top of Numpy,provides data structures or objects  like Series or DataFrame.

## Why do we use Pandas?
1. Easy data cleaning and transformation.
2. Efficient handling  of large datasets. 
3. Integration with other Python libraries like "Matplotlib" and  "Seaborn" for visualization.

## Main data structures or objects in Pandas:
1. Series. (1 Dimensional)
2. DataFrame. (2 Dimensional)

```python
df2.dropna() # It will delete those datapoints (.i.e. rows) where entries are missing (Rows where "NaN" is present)
# See below 0th row , 2nd row and 4th row, etc. are deleted.
```

```python
# df2.fillna(0,inplace=True)  Never use this inplace=True, which leads to the modification of original DataFrame(data).
df2.fillna(0)  # "NaN" entries are replaced by "0"
```

```python
df.rename(columns = {"SepalLengthCm":"SL"}) # It renames the column namely "SepalLengthCm" to the "SL". It
# will also print the DataFrame with modified column name. Note that this "rename() function" or any "other functions/methods"
#  that applied to DataFrame, will not change the original data. It always make its copy with corresponding changes unless 
# you have not used "inplace=True". 
```

```python
df.info() # see this is the original data which have not changed. Note the column name "SepalLengthCm" remain intact.
# Note below datatypes are given for columns-entries, which is by default given by "Pandas". 
# Example, float64 -> floating point number.We can change this datatype of particular column by using 
# method  ".astype(#name of datatype#)"  See below
```

```python
df["SepalLengthCm"] = df["SepalLengthCm"].astype(int) # This will be inplace modification, that means In original data 
# we will see "SepalLengthCm" column have integer values rather than floating values.
df.info()
```

```python
print(df)
```

```python
df["SepalLengthCm"] = df["SepalLengthCm"].astype(str)
df.info() 
```

```python
(df["SepalLengthCm"][1]) # see this will give you entry in "column-> "SepalLengthCm" and row -> "1"." This entry is string
# which will be displayed in single inverted comma''. In past, Pandas store datatypes except some of them as python object
# that's why it is displayed as object in "df.info()" If you want string to be displaqyed -> see below code
```

```python
df["SepalLengthCm"] = df["SepalLengthCm"].astype("string")
df.info()
```

```python
len(df) # display the number of rows
```

```python
df['zeros'] = [0 for i in range(len(df))] # See we have added new column namely "zeros" with "0" entries.
df # this will print the data, no need of print()
```

```python
def f(a):
    return a+1

df["zeros+1"] = df["zeros"].apply(f) # It will create the new column namely "zeros+1".
# Corres. entry of "zeros" column as the argument for function "f" and returned value of f 
# will be the corresponding entry of "zeros +1 " column.
df
```

```python
df.to_csv("export.csv",index=False) # This will export the same data file (make a copy of it). If we do "not" use "index=False"
# it will give us the default column of index.
```

```python
df1 = pd.DataFrame({"names":["Akash", "Raghu"],"Marks":[15,67]})
df1
```

```python
df2 = pd.DataFrame({"names":["Akash", "Raghu"],"RollNo":[5,7]})
df2
```

```python
pd.merge(df1,df2, on="names") # It will merge the both data based on the label "names",for merging common entries must be there
# in the column label "names",otherwise the uncommon entries in the label "names" will be discarded. (see below)
```

```python
df1 = pd.DataFrame({"names":["Akash", "Raghu", "Harry"],"RollNo":[5,7,3]})
df2 = pd.DataFrame({"names":["Akash", "Raghu"],"Marks":[15,67]})
pd.merge(df1,df2,on="names")
```
